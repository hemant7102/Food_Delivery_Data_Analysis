```python
import sqlite3

conn = sqlite3.connect(r"C:\Users\Hemant\Downloads\restaurants.db")
cursor = conn.cursor()

with open(r"C:\Users\Hemant\Downloads\restaurants.sql", "r", encoding="utf-8") as f:
    sql_script = f.read()

cursor.executescript(sql_script)
conn.commit()


pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
)

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()

```

```python
restaurants.head()
```

```python
# Merging

merged1 = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"
)
```

```python
merged1.head()
```

```python
final_df = pd.merge(
    merged1,
    restaurants,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.head()
```

```python
final_df.to_csv("final_food_delivery_dataset.csv",index=False)
```

## 4>. Which restaurants rating range generated the highest total revenue

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0-3.5', '3.6-4.0', '4.1-4.5', '4.6-5.0']

df['rating_range'] = pd.cut(df['rating'], bins=bins, labels=labels, include_lowest=True)
```

```python
rating_revenue = df.groupby('rating_range')['total_amount'].sum().sort_values(ascending=False)
rating_revenue
```

```python
top_rating_range = rating_revenue.head(1)
top_rating_range
```
